# K-nearest neighbors

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

A seguir vamos importar nossos dados e dar uma espiada nas primeiras linhas dele

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
X = df[['Pregnancies', 'Glucose', 'BloodPressure', 'BMI', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = df['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

Temos uma acurácia de cerca de 70%, que pode ser melhorada trocando o modelo ou com técnicas auxiliares, mas como escolher nosso modelo é algo que veremos futuramente após ter uma breve introdução a alguns modelos

```python
# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=df['Outcome'])
disp.plot(cmap=plt.cm.Blues)
plt.show()
```
